```python
from Bio import SeqIO
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import src.sequence_utils as su
from importlib import reload
import os

sns.set_style("darkgrid")

fasta1 = os.listdir("data")
recs = list(SeqIO.parse("data/sample-NC_000913.fasta", "fasta")) #This is a list of sequences recorded inside of the fasta file
seq1 = recs[0]

print("Imports done")
```

Count how many of each nucleotide (A, C, G, T) are in seq.
seq: a Bio.Seq.Seq or Python string of A/C/G/T.
Returns a dict like {'A': 123, 'C': 98, 'G': 110, 'T': 105}.

```python

def base_counts(seq):

    return {nt: seq.count(nt) for nt in "ACGT"}

#NOTE TO SELF: How to write base-counts function like that to generate a dict???
#Using our above function, we generated a dictionary of each nucleotide and their count
counts = base_counts(seq1.seq) 

#We can now use dictionary functions to look at our new dict
print("Counts dict:", counts)
print("Sum of counts:", sum(counts.values()))
print("Seq length:", len(seq1.seq))

    
```

```python
char_counts = {}

for letter in "ACGT":
    char_counts[letter] = "ACGT".count(letter)

print(char_counts)

#--------------------------

even_squares = [ n*n for n in range(10) if n % 2 == 0 ]

print(even_squares)
```

```python

# items() is a method of dicts. Returns a view of dict as (key, value) pairs in the form of two length tuples within a list
# zip() returns a zip object, which is an iterator of tuples. Each tuple contains one element from each of the input iterables, paired by position.
# Input pairs: [('A',10),('C',5),('G',7),('T',8)]
# zip(*…) → [('A','C','G','T'), (10,5,7,8)]

bases, freqs = zip(*counts.items())

#We utilize matplotlib as plt here and their commands.

#Figure() reserves a blank canvas, with arguments that look like "property=specification" you can edit.
#For simple plots, we organize our data into simple tutples/lists for the x/y axes 
plt.figure(figsize=(6,5))
plt.bar(bases, freqs,
        color="green",
        width =0.5
       )

#Decorate with titles, labels, legend and tweak layout
plt.title("Nucleotide Frequencies of E coli str. K-12 substr. MG1655")

plt.xlabel("Base")


plt.ylabel("Count")
plt.yscale("log")

plt.tight_layout()
plt.style.use("seaborn-v0_8-whitegrid")


#Once done with instantiation, use show() and savefig(name, dpi) for viewing. The path can be customized in the name argument
plt.show()
plt.savefig("../results/nuc_freq.png", dpi=300)

```

```python
#GC Content analysis

chunkSize = 75000
step = 25000

positions = []
gc_values = []

def gc_contents(seq):

    gc = 0
    cc = 0

    for base in seq:
        if base == "G":
            gc+=1
        elif base == "C":
            cc+=1

    return 100 * ((gc+cc)/len(seq))

for i in range(0, len(seq1)-chunkSize+1,step):
    positions.append(i)
    gc_values.append(gc_contents(seq1[i:i+chunkSize]))
    
fig, ax = plt.subplots(figsize=(8,4), layout="constrained")

ax.plot(positions, gc_values,
         linestyle='-',
         marker='o',
         markersize=4,
         color='navy',
        linewidth=0.7,
        alpha=0.6
        )

ax.set(
    xlabel="Start Position (bp)",
    ylabel="GC Content (%)",
    title="Sliding-Window GC Curve of E coli str. K-12 substr. MG1655"
)

ax.grid(True)
ax.xaxis.set_major_locator(MaxNLocator(nbins=10))
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))



plt.savefig("results/sliding-window-GC.png", dpi=300)
plt.show()
```
