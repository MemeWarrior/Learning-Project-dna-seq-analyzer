```python
from Bio import SeqIO
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
from importlib import reload
from src.sequence_utils import DATA_DIR, sliding_gc

sns.set_style("darkgrid")

recs = list(SeqIO.parse("data/sample-NC_000913.fasta", "fasta")) #This is a list of sequences recorded inside of the fasta file
recs2 = list(SeqIO.parse("data/sample-rtropici.fasta", "fasta"))

seq1 = recs[0]
seq2 = recs2[0]

print("Imports Finished")
```

```python
from src.sequence_utils import sliding_gc

positions, gc_values = sliding_gc(seq2, 100000,25000)

fig, ax = plt.subplots(figsize=(8,4), layout="constrained")

ax.plot(positions, gc_values,
         linestyle='-',
         marker='o',
         markersize=4,
         color='navy',
        linewidth=0.7,
        alpha=0.6
        )

ax.set(
    xlabel="Start Position (bp)",
    ylabel="GC Content (%)",
    title="Sliding-Window GC Curve of " + "Name", 
)

ax.grid(True)
ax.xaxis.set_major_locator(MaxNLocator(nbins=10))
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))



plt.savefig("results/sliding-window-GC.png", dpi=300)
plt.show()
```
